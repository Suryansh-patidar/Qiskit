```python
coords = [math.pi/2,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

```python
coords = [math.pi/2,math.pi,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

* Looking at the first and last two states, we can say that first two are opposite to each other, same as the last two. Also we are just rotating the first two by 90 degrees in the anticlockwisedirection to get the last two


* The directions can be specified by just looking at the values of 'theta' and 'phi'. Also if you know that '|0>'th state is basically z-axis and '|1>' is the opposite of that, x or '-x' axes can be thought of as the superposition of '|0>'th and '|1>'st state with equal probabilities, and if you want to know about the signs, just look at values of 'theta'.
Another way to look at it is just imagining the values 'theta' and 'phi' as standard angles w.r.t z-axis and x-axis respectively. It gives more feel to it
